```python
from ipywidgets import interact
```

```python
def plot_sine(a,k):
    x = linspace(0,1,1001)
    y = a*sin(pi*k*x)
    plot(x,y,'b')
    ylim(-5,5)
    title('%s sin(%s \pi x)' % (a,k))
```

```python
plot_sine(3,5)
```

```python
interact(plot_sine,a=(1,4,1),k=(1,10,1))
```
